This is an example of **how not** to use Deep Learning

```python
from six.moves.urllib.request import urlretrieve

filename = "BlogFeedback.zip"
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00304/BlogFeedback.zip"
urlretrieve(url,filename)

import zipfile

zip_ref = zipfile.ZipFile(filename, 'r')
zip_ref.extractall("blogData")
zip_ref.close()
```

```python
import pandas as pd
import numpy as np

df_train = pd.read_csv("blogData/blogData_train.csv",header=None)
x_train = df_train.values[:,:-1]
y_train = df_train.values[:,-1]
```

```python
x_train = df_train.values[:,:-1]
y_train = df_train.values[:,-1]
```

```python
df_train.head(10)
```

```python
print(df_train.shape)
```

```python
import glob
test_files = glob.glob("blogData/blogData_test*.csv")

dfs = []
for file in test_files:
    dfs.append(pd.read_csv(file,header=None))
    
df_test = pd.concat(dfs)    

x_test = df_test.values[:,:-1]
y_test = df_test.values[:,-1]
```

```python
df_test.head(20)
```

```python
df_test.shape
```

```python
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler

# Scale xs and ys to lie between 0 and 1
scaler_x = MinMaxScaler()
x_train = scaler_x.fit_transform(x_train)
x_test = scaler_x.transform(x_test)

scaler_y = MinMaxScaler()
y_train = scaler_y.fit_transform(y_train[:,None])
y_test = scaler_y.transform(y_test[:,None])

lin_reg = linear_model.LinearRegression()
lin_reg.fit(x_train,y_train)
print(lin_reg.coef_)
```

```python
y_train_pred = lin_reg.predict(x_train)
```

View the predicted value against the true value

```python
import numpy as np

np.hstack([y_train_pred,y_train])
```

```python
e = y_train - y_train_pred
r2 = 1 - np.var(e)/np.var(y_train)
r2
```

Check the same thing against test case

```python
y_test_pred = lin_reg.predict(x_test)
e = y_test - y_test_pred
r2 = 1 - np.var(e)/np.var(y_test)
r2
```

Use one more linear method

```python
lin_reg = linear_model.Lasso(alpha=1e-5)
lin_reg.fit(x_train,y_train)
print(lin_reg.coef_)
```

```python
lin_reg.score(x_train,y_train)
```

```python
lin_reg.score(x_test,y_test)
```

## Deep Learning model

```python
from keras.models import Sequential, Model
from keras.layers import Dense, Embedding, Dropout
from keras.regularizers import l2

import theano
theano.config.compute_test_value = 'ignore'

model = Sequential()

model.add(Dense(input_dim=x_train.shape[1],output_dim=500,activation="relu"))
# model.add(Dense(10,activation="relu"))
model.add(Dense(1))
model.compile(optimizer='adagrad', loss='mean_squared_error')
#                     metrics=['mean_squared_error'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, nb_epoch=5, batch_size=128,verbose=1)
```

```python
y_train_pred = model.predict(x_train,batch_size=1024,verbose=1)
e = y_train - y_train_pred
r2 = 1 - np.var(e)/np.var(y_train)
r2
```

```python
y_test_pred = model.predict(x_test)
e = y_test - y_test_pred
r2 = 1 - np.var(e)/np.var(y_test)
r2
```
